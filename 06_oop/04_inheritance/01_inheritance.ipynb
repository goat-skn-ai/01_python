# 상속 Inheritance
- 부모클래스 속성/메소드를 자식클래스에서 선언없이 물려받아 사용하는 것
- 중복된 속성/메소드를 재작성할 필요가 없다.
- 자식클래스를 부모클래스로 묶어 관리할 수 있다.
- 파이썬 다중상속 지원

```python
class Person:
    def __init__(self, name: str, phone: str) -> None:
        self.__name = name
        self.__phone = phone

    def hello(self):
        print(f'안녕하세요, 저는 {self.__name}입니다. 제 연락처는 {self.__phone}이에요.')

    def get_name(self):
        return self.__name

    def get_phone(self):
        return self.__phone

    def __str__(self):
        return f'Person(name={self.__name}, phone={self.__phone})'


person = Person('홍길동', '01012341234')
print(person)
```

```python
# 부모클래스를 ()안에 작성
class Employee(Person):

    def __init__(self, name: str, phone: str, company: str) -> None:
        # 부모생성자를 호출
        super().__init__(name, phone)
        self.__company = company

    def get_company(self):
        return self.__company

    # 메소드 재작성 (method overriding)
    # - 물려받은 메소드를 고쳐쓰고 싶을 때 작성
    def hello(self):
        super().hello()  # 부모메소드 호출
        print(f'저는 {self.__company}에서 일하고 있습니다.')

    def __str__(self):
        return f'Employee(name={self.get_name()}, phone={self.get_phone()}, company={self.__company})'


emp = Employee('신사임당', '01012345678', 'SK Network')
print(emp.get_name(), emp.get_phone(), emp.get_company())

emp.hello()

print(emp)
```

```python
# Dev
# - Person 상속
# - langs: list[str]
# - add_lang(lang: str) 언어 추가
# - remove_lang(lang: str) 언어 삭제 (언어)

class Dev(Person):

    def __init__(self, name: str, phone: str, *langs: tuple[str]) -> None:
        super().__init__(name, phone)
        self.__langs = list(langs)

    def hello(self):
        super().hello()
        print(f'저는 {self.__langs} 언어를 사용합니다.')

    def add_lang(self, lang: str):
        if lang not in self.__langs:
            self.__langs.append(lang)
            print(f'{lang}이 추가되었습니다.')
        else:
            print(f'{lang}은 이미 추가되어 있습니다.')

    def remove_lang(self, lang: str):
        if lang in self.__langs:
            self.__langs.remove(lang)
            print(f'{lang}을 삭제했습니다.')
        else:
            print(f'{lang}은 존재하지 않습니다.')

    def __str__(self):
        return f'Dev(name={self.get_name()}, phone={self.get_phone()}, langs={self.__langs})'

dev = Dev('홍길동', '01012345678', 'python', 'java', 'c++')
dev.hello()
dev.add_lang('javascript')
print(dev)
dev.remove_lang('javaaaaaaa')
dev.remove_lang('java')
```
