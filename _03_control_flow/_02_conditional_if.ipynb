# 조건문 if

- 조건문은 특정 조건에 따라 실행할 코드를 선택하는 문법이다.
- `if`는 조건이 참일 때만 실행한다.
- `if...else`는 둘 중 하나를 반드시 실행한다.
- `if...elif...else`는 여러 조건 중 하나를 선택한다.

## 1. if

`if`는 특정 조건에서만 실행되어야 하는 코드가 있을 때 사용한다.

예를 들어 시험 점수가 90점 이상일 때만 커피 쿠폰을 지급할 수 있다.

```python
def give_coupon(score: int) -> None:
    if score >= 90:
        print('☕☕☕ 커피 쿠폰을 드립니다.')

    print('😊수고하셨습니다.😊')

print('--- if 예제 ---')
give_coupon(95)
give_coupon(70)
```

## 2. if...else

`if...else`는 조건이 참인 경우와 거짓인 경우를 모두 처리해야 할 때 사용한다.

예를 들어 점수가 60점 이상이면 합격, 그렇지 않으면 불합격으로 나눌 수 있다.

```python
def pass_or_fail(score: int) -> None:
    if score >= 60:
        print('합격입니다.')
    else:
        print('불합격입니다.')

    print('😊수고하셨습니다.😊')

print('--- if...else 예제 ---')
pass_or_fail(80)
pass_or_fail(55)
```

## 3. if...elif...else

`if...elif...else`는 고려해야 하는 조건이 여러 개인 경우 사용한다.

학점 계산처럼 여러 구간 중 하나를 선택해야 할 때 적합하다.

```python
def get_grade(score: int) -> str:
    """시험 점수에 따라 학점 등급을 반환한다."""
    if not (0 <= score <= 100):
        raise ValueError('유효한 점수를 입력해주세요...')

    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    elif score >= 60:
        grade = 'D'
    else:
        grade = 'F'

    return grade

print('--- if...elif...else 학점 예제 ---')
for score in [95, 85, 75, 65, 55]:
    grade = get_grade(score)
    print(f'{score}점 -> {grade}등급')
```

## 4. 입력값 유효성 검사

사용자 입력은 항상 올바르다고 가정하면 안 된다. 점수처럼 범위가 정해진 값은 조건문으로 유효성을 먼저 확인하는 습관이 좋다.

```python
print('--- 유효성 검사 예제 ---')

for score in [100, -1, 101]:
    try:
        grade = get_grade(score)
        print(f'{score}점 -> {grade}등급')
    except ValueError as error:
        print(f'{score}점 -> 오류:', error)

# 직접 입력 실습용
# score: int = int(input('시험점수 입력: '))
# print(get_grade(score))
```
