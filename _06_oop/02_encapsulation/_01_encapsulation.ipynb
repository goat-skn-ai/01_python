# 캡슐화
- 속성/메서드를 하나의 클래스에 함께 선언
- private 속성(변수) 선언해서 외부에서 직접 접근을 막고, 제공된 메소드를 통해서 사용하도록 강제
  - private 변수 선언 방법 : 변수명 앞에 __를 붙인다.
- 캡슐화 원칙을 잘 지켰다면, 주요한 데이터가 외부에 노출되지 않은 은닉화 효과까지 얻게된다.

```python
class Account:

    def __init__(self, username, balance):
        # private 접근제한 처리
        self.__username = username
        self.__balance = balance

    def info(self):
        print(f'{self.__username}님의 현재잔액은 {self.__balance}원 입니다.')

    def deposit(self, money):
        """
        입금메소드
        - money 유효성 검사
        - 오염된 입금처리 방지
        """
        if money <= 0:
            print('입금액이 유효하지 않습니다.')
        else:
            self.__balance += money
            print('입금이 정상처리 되었습니다.')

    def withdraw(self, money):
        """출금메소드"""
        if money <= 0:
            print('출금액이 유효하지 않습니다.')
        elif money > self.__balance:
            print('잔액이 부족합니다.')
        else:
            self.__balance -= money
            print('출금이 정상처리 되었습니다.')
```

```python
# private 속성 접근 확인
# 홍길동의 계좌관리 (필드 직접 사용)
acc = Account('honggd', 1_000_000)
print('acc 객체:', acc)
acc.info()

try:
    # __balance는 외부에서 balance라는 이름으로 직접 읽을 수 없다.
    acc.balance = acc.balance - 700_000
except AttributeError as error:
    print('private 속성 직접 접근 결과:', error)

acc.info()
```

## Python private과 name mangling

Python에서 `__balance`처럼 언더스코어 두 개로 시작하는 속성은 외부에서 그대로 접근하기 어렵게 이름이 바뀐다.
이를 name mangling이라고 한다.

다만 이것은 보안을 위한 완전한 접근 제한이라기보다, 클래스 외부에서 직접 사용하지 말라는 강한 신호에 가깝다.


```python
print('--- name mangling 확인 ---')

acc = Account('honggd', 1_000_000)

print('acc.__dict__:', acc.__dict__)
print('name mangling된 balance 속성:', acc._Account__balance)

# 아래처럼 접근할 수는 있지만, 캡슐화 원칙을 깨므로 권장하지 않는다.
# acc._Account__balance = 0
```

```python
# 홍길동의 계좌관리 (메서드 이용, 캡슐화)
acc = Account('honggd', 1_000_000)
print(acc)
acc.info()

# 입금처리
acc.deposit(500_000)
acc.deposit(-500_000)
acc.info()

# 출금처리
acc.withdraw(700_000)
acc.withdraw(-700_000)
acc.withdraw(40_000_000)
acc.info()
```
