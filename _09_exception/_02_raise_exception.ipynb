# 예외처리 흐름
raise 예외객체를 사용해서 명시적으로 예외를 던질수 있다.
이를 통해 코드 처리흐름을 명확히 전달할 수 있다.

```python
# raise : 예외를 직접 던지는 구문
def foo():
    # raise NotImplementedError # 예외메세지 없는 경우 클래스만 작성가능
    raise NotImplementedError("foo함수 어쩌고저쩌고한 기능 구현예정")
foo()
```

```python
# if...else 흐름
class AdultGame:
    def __init__(self):
        pass

    def start(self):
        print('성인용 게임을 시작합니다...')

age = int(input("나이입력 : "))
if age >= 20:
    AdultGame().start()
    # game = AdultGame()
    # game.start()
else:
    print("미성년자는 게임을 이용하실 수 없습니다.")
```

```python
# AdultGame 클래스에 나이검사 기능 추가
# 커스텀 예외클래스 (의도 분명히 전달)
class UnderAgeError(Exception):
    pass

class AdultGame2:
    def __init__(self, age):
        if age < 20:
            raise UnderAgeError(f'미성년자는 이용할 수 없습니다: {age}')

    def start(self):
        print('성인용 게임을 시작합니다...')


try:
    age = int(input("나이 입력: "))
    AdultGame2(age).start()
except UnderAgeError as e:
    print(e)
```

```python
# 사용자로부터 파일명을 입력받고, 해당파일의 내용을 읽어서 출력
# - 만약 해당 파일이 존재하지 않으면, 사용자 입력값으로 파일을 직접 생성
# 1. 파일명 입력받기
# 2. 해당파일 열어서 읽기

try:
    filename: str = input("읽어올 파일명 입력: ")
    with open(filename, 'r', encoding='utf-8') as f:
        print(f.read())
except FileNotFoundError as e:
    print(f'{filename} 파일은 존재하지 않습니다. 새로 만들어드리겠습니다.')
    content = input("내용을 입력해주세요 : ")
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(content)
        print('해당 내용을 성공적으로 저장했습니다.')

print('😊😊😊이용해 주셔서 감사합니다😊😊😊')

```
