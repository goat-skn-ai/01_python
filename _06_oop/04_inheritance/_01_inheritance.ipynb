# 상속 Inheritance
- 부모클래스 속성/메서드를 자식클래스에서 선언없이 물려받아 사용하는 것
- 중복된 속성/메서드를 재작성할 필요가 없다.
- 자식클래스를 부모클래스로 묶어 관리할 수 있다.
- 파이썬 다중상속 지원

```python
class Person:
    def __init__(self, name: str, phone: str) -> None:
        self.__name = name
        self.__phone = phone

    def hello(self):
        print(f'안녕하세요, 저는 {self.__name}입니다. 제 연락처는 {self.__phone}이에요.')

    def get_name(self):
        return self.__name

    def get_phone(self):
        return self.__phone

    # __str__(): print(객체)를 했을 때 어떤 문자열을 보여줄지 정하는 메서드
    def __str__(self):
        return f'Person(name={self.__name}, phone={self.__phone})'


person = Person('홍길동', '01012341234')
print(person)
```

```python
# 부모클래스를 ()안에 작성
class Employee(Person):

    def __init__(self, name: str, phone: str, company: str) -> None:
        # 부모생성자를 호출
        super().__init__(name, phone)
        self.__company = company

    def get_company(self):
        return self.__company

    # 메소드 재작성 (method overriding)
    # - 물려받은 메소드를 고쳐쓰고 싶을 때 작성
    def hello(self):
        super().hello()  # 부모메소드 호출
        print(f'저는 {self.__company}에서 일하고 있습니다.')

    def __str__(self):
        return f'Employee(name={self.get_name()}, phone={self.get_phone()}, company={self.__company})'


emp = Employee('신사임당', '01012345678', 'SK Network')
print(emp.get_name(), emp.get_phone(), emp.get_company())

emp.hello()

print(emp)
```

```python
# Dev
# - Person 상속
# - langs: list[str]
# - add_lang(lang: str) 언어 추가
# - remove_lang(lang: str) 언어 삭제 (언어)

class Dev(Person):

    def __init__(self, name: str, phone: str, *langs: tuple[str]) -> None:
        super().__init__(name, phone)
        self.__langs = list(langs)

    def hello(self):
        super().hello()
        print(f'저는 {self.__langs} 언어를 사용합니다.')

    def add_lang(self, lang: str):
        if lang not in self.__langs:
            self.__langs.append(lang)
            print(f'{lang}이 추가되었습니다.')
        else:
            print(f'{lang}은 이미 추가되어 있습니다.')

    def remove_lang(self, lang: str):
        if lang in self.__langs:
            self.__langs.remove(lang)
            print(f'{lang}을 삭제했습니다.')
        else:
            print(f'{lang}은 존재하지 않습니다.')

    def __str__(self):
        return f'Dev(name={self.get_name()}, phone={self.get_phone()}, langs={self.__langs})'

dev = Dev('홍길동', '01012345678', 'python', 'java', 'c++')
dev.hello()
dev.add_lang('javascript')
print(dev)
dev.remove_lang('javaaaaaaa')
dev.remove_lang('java')
```

## isinstance(), issubclass(), MRO

상속 관계에서는 객체가 어떤 클래스의 인스턴스인지, 클래스가 어떤 부모를 상속하는지 확인할 수 있다.

- `isinstance(객체, 클래스)`: 객체가 해당 클래스 또는 하위 클래스의 인스턴스인지 확인
- `issubclass(자식클래스, 부모클래스)`: 클래스 사이의 상속 관계 확인
- MRO(Method Resolution Order): 메서드를 찾는 순서


```python
print('--- 상속 관계 확인 ---')

print('isinstance(emp, Employee):', isinstance(emp, Employee))
print('isinstance(emp, Person):', isinstance(emp, Person))
print('isinstance(dev, Person):', isinstance(dev, Person))

print('issubclass(Employee, Person):', issubclass(Employee, Person))
print('issubclass(Dev, Person):', issubclass(Dev, Person))

print('Employee.mro():', Employee.mro())
```

## Python의 오버라이딩과 오버로딩 주의

Python은 자식 클래스에서 부모 클래스의 메서드를 다시 정의하는 오버라이딩은 자연스럽게 지원한다.
다만 Java처럼 같은 이름의 메서드를 매개변수 조합별로 여러 개 선언하는 메서드 오버로딩은 지원하지 않는다.

같은 클래스 안에서 같은 이름의 메서드를 다시 선언하면, 뒤에 선언한 메서드가 앞의 메서드를 덮어쓴다.


```python
print('--- Python 메서드 오버로딩 주의 ---')

class Calculator:
    def add(self, a):
        return a

    def add(self, a, b):
        return a + b


calculator = Calculator()

# print('calculator.add(10):', calculator.add(10))  # TypeError: b가 필요하다.
print('calculator.add(10, 20):', calculator.add(10, 20))
```
