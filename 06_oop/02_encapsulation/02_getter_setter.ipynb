## Getter/Setter 메소드
캡슐화된 클래스에서 속성을 제어하기 위한 메소드로써 기본적으로 다음 두 유형이 있다.
1. getter 속성의 값을 읽어내기 위한 메소드
2. setter 속성의 값을 설정하기 위한 메소드

```python
# class명은 관례적으로 파스칼케이싱을 따른다.
class MyClass:

    def __init__(self, value):
        self.__value = value

    # 읽기 가능한 속성인 경우, getter를 선언해둔다.
    def get_value(self):
        return self.__value

    # 쓰기 허용 속성이라면, setter를 선언해둔다.
    def set_value(self, value):
        self.__value = value


my_clz = MyClass(10)

# 값 읽기
# my_clz.__value # AttributeError: 'MyClass' object has no attribute '__value'
print(my_clz.get_value())

# 값 쓰기
# my_clz.__value = 123
my_clz.set_value(20)

print(my_clz.get_value())
```
