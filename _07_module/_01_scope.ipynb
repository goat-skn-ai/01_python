# 유효범위 Scope

변수의 유효범위(scope)는 변수에 접근할 수 있는 범위를 의미한다.

- 전역 변수(global variable): 함수 블록 밖에서 선언되어 여러 위치에서 접근할 수 있는 변수
- 지역 변수(local variable): 함수 내부에서 선언되어 해당 함수 안에서 주로 사용하는 변수
- Python의 `for`, `if` 블록은 Java의 중괄호 블록처럼 별도의 지역 스코프를 만들지 않는다.


## 1. 전역 변수와 지역 변수

함수 밖에서 선언된 `a`는 전역 변수이고, 함수 안에서 선언된 `b`는 지역 변수이다.
`b`는 `foo()` 함수 내부에서만 사용할 수 있다.


```python
a = 100  # 전역 변수


def foo():
    print('foo()에서 참조한 전역변수 a:', a)

    b = 200  # 지역 변수
    print('foo()에서 선언한 지역변수 b:', b)


foo()

# print('foo() 밖에서 지역변수 b:', b)  # NameError: name 'b' is not defined
# print('foo.b 함수 객체의 속성 b:', foo.b)  # AttributeError: 'function' object has no attribute 'b'
```

## 2. for, if 블록과 스코프

Python에서는 `for`, `if` 블록이 별도의 지역 스코프를 만들지 않는다.
따라서 블록 안에서 만들어진 이름을 블록 밖에서도 참조할 수 있다.


```python
for i in range(3):
    print('for문의 현재 반복 변수 i:', i)

print('for문 종료 후 남아 있는 i:', i)

if True:
    k = 100

print('if 블록에서 선언한 k:', k)
```

## 3. 이름 탐색 우선순위

같은 이름의 변수가 여러 범위에 있을 때는 더 가까운 범위의 변수가 우선된다.
함수 안에서 같은 이름으로 대입하면 전역 변수를 바꾸는 것이 아니라, 지역 변수를 새로 만든다.


```python
x = 100


def bar():
    x = 200
    print('bar() 안의 지역변수 x:', x)


bar()
print('bar() 호출 후 전역변수 x:', x)
```

## 4. 전역 변수와 지역 변수가 함께 사용되는 경우

아래 예제에서 `koo()` 함수 안의 `glo`는 지역 변수이고, 함수 밖의 `glo`와는 다른 이름 공간에 있다.


```python
glo = 10  # 전역 변수
bal = 5   # 전역 변수


def koo(a):
    glo = a + 10  # 지역 변수 glo
    print('koo() 안의 지역변수 glo:', glo)
    b = a + glo + bal  # 지역 a + 지역 glo + 전역 bal
    return b


print('koo(10) 반환값:', koo(10))
print('koo(10) 호출 후 전역변수 glo:', glo)
print('koo(bal) 반환값:', koo(bal))
print('전역변수 bal:', bal)
```

## 5. immutable, mutable 객체 전달

Python은 함수에 값을 전달할 때 객체에 대한 참조를 전달한다.
다만 immutable 객체는 내부 값을 직접 바꿀 수 없고, mutable 객체는 내부 값을 바꿀 수 있어 결과가 다르게 보인다.

- immutable: `int`, `float`, `bool`, `tuple`, `str`
- mutable: `list`, `dict`, `set`


```python
def func1(a):
    a = 200


a = 100
func1(a)
print('func1(a) 호출 후 immutable 변수 a:', a)


def func2(b):
    b[1] *= 100


b = [1, 2, 3]
func2(b)
print('func2(b) 호출 후 mutable 리스트 b:', b)
```
