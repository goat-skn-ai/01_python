## 파일 쓰기

```python
# open(파일경로, 모드, **options)
# - 경로 : 현재경로 ./ 생략가능
f = open('./helloworld.txt', 'w', encoding='utf-8')
print(f, type(f))

# 지정된 파일이 있다면 덮어쓰기, 없다면 파일 생성
f.write('''Hello
World
😘☑️🍖''')

# 사용후 반드시 반환 (메모리 해제)
f.close()
```

```python
# with절 파일열기/닫기
# - with절을 이용하면 자동 close 된다
with open('./helloworld.txt', 'w', encoding='utf-8') as f:
    f.write('''🍔
🍖
🍕
☕''')
    # 자동반환
```

## 파일 읽기

```python
# 파일 전체 내용 읽기
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    text = f.read() # 전부 읽어오기
    print(text)
```

```python
# 한 줄씩 읽기
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        print(i, line, end='')
```

```python
# 파일 이어쓰기
# - mode: a(append)
with open('helloworld.txt', 'a', encoding='utf-8') as f:
    f.write('''🤣🤣
😂😂
😊😊
''')

with open('helloworld.txt', 'r', encoding='utf-8') as f:
    text = f.read() # 전부 읽어오기
    print(text)
```

```python
# 파일복사
# 1. 파일읽기
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# 2. 파일쓰기
with open('helloworld_copy.txt', 'w', encoding='utf-8') as f:
    f.write(text)
```

```python
# 이미지 복사하기
with open("1_.png", 'rb') as f:
    #print(f.read())
    image = f.read()

with open('1_copy.png', 'wb') as f:
    f.write(image)
```

## dict-csv 읽고 쓰기

```python
users = [
    {"id": 100, "name": "홍길동", "email": "honggd@naver.com"},
    {"id": 200, "name": "신사임당", "email": "sinsa@naver.com"},
    {"id": 300, "name": "논개", "email": "nondog@gmail.com"},
]
```

```python
# csv.DictWriter: 딕셔너리의 key를 CSV 컬럼명과 매칭해서 파일에 쓰는 도구
import csv

with open('users.csv', 'w', encoding="utf-8", newline='') as f:
    writer = csv.DictWriter(f, fieldnames=["id", "name", "email"]) # CSV의 컬럼 순서를 지정
    writer.writeheader() # CSV 파일의 첫 줄에 헤더를 작성

    # 여러줄 쓰기
    # writer.writerows(users)

    # 한줄씩 쓰기
    for user in users:
        writer.writerow(user)
```

```python
with open('users.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)

    # for row in reader:
    #     print(row, type(row))

    users = [row for row in reader]
    for user in users:
        print(user)
```
