# Type Hint

- Python은 **동적 타이핑(dynamic typing)** 언어이다.
- 변수 이름 자체에 자료형이 고정되는 것이 아니라, 변수가 가리키는 객체가 타입을 가진다.
- **type hint**는 코드 작성 시 IDE와 정적 분석 도구가 타입을 추론하고 경고할 수 있도록 돕는 표시이다.
- 중요한 점은 type hint가 Python 런타임에서 타입을 강제로 막아 주지는 않는다는 것이다.

## 1. 동적 타이핑

Python에서는 같은 변수 이름에 서로 다른 타입의 객체를 다시 대입할 수 있다.

- `x = '안녕'`일 때 `x`는 문자열 객체를 가리킨다.
- `x = 123`으로 다시 대입하면 `x`는 정수 객체를 가리킨다.
- 타입은 변수 이름이 아니라, 변수 이름이 가리키는 객체가 가진다.

````python
print('--- 동적 타이핑 ---')

x = '안녕'
print('문자열 대입 후 x:', x)
print('문자열 대입 후 type(x):', type(x))

x = 123
print('정수 재대입 후 x:', x)
print('정수 재대입 후 type(x):', type(x))
````

## 2. 함수 정의

타입 힌트가 없는 함수도 Python에서는 정상적으로 정의하고 실행할 수 있다.

다만 매개변수에 어떤 타입을 넣어야 하는지 코드만 보고 알기 어렵기 때문에, 큰 코드에서는 type hint가 가독성과 유지보수에 도움이 된다.

````python
print('--- 함수 정의 ---')

def act(x):
    """
    전달받은 값을 출력하는 예시 함수
    :param x: 출력할 값
    :return: None
    """
    print('act 함수에 전달된 x:', x)

act('안녕')
act(123)
````

## 3. 변수 Type Hint

변수 뒤에 `: 타입`을 붙이면 타입 힌트를 작성할 수 있다.

```python
변수명: 타입 = 값
```

하지만 type hint는 실행 중 타입을 강제하지 않는다. 아래 예시처럼 `str`이라고 힌트를 작성해도, 런타임에서는 `int` 재대입을 직접 막지 않는다.

````python
print('--- 변수 type hint ---')

greeting: str = 'Hello'
print('처음 greeting:', greeting, type(greeting))

# type hint를 str로 작성했지만, 런타임에서는 int 재대입을 막지 않는다.
greeting = 123
print('int 재대입 후 greeting:', greeting, type(greeting))

n: float = 123.456
score: int = 90
has_coupon: bool = True

print('n:', n, type(n))
print('score:', score, type(score))
print('has_coupon:', has_coupon, type(has_coupon))
````

## 4. 컬렉션 Type Hint

`list`, `tuple`, `dict`, `set` 같은 컬렉션에도 type hint를 작성할 수 있다.

- `list[int]`: 정수를 담는 리스트
- `tuple[int, str, str]`: 첫 번째는 정수, 두 번째와 세 번째는 문자열인 튜플
- `dict[str, str | int]`: key는 문자열, value는 문자열 또는 정수인 딕셔너리
- `set[str]`: 문자열을 담는 집합

````python
print('--- 컬렉션 type hint ---')

nums: list[int] = [1, 2, 3]
user: tuple[int, str, str] = (123, '홍길동', 'honggd')
info: dict[str, str | int] = {'name': '신사임당', 'age': 42}
chars: set[str] = {'a', 'b', 'c'}

print('nums:', nums, type(nums))
print('user:', user, type(user))
print('info:', info, type(info))
print('chars:', chars, type(chars))
````

## 5. 함수 Type Hint

함수의 매개변수와 반환값에도 type hint를 작성할 수 있다.

```python
def 함수명(매개변수: 타입) -> 반환타입:
    ...
```

이 표기는 함수 사용법을 문서처럼 보여 주고, IDE나 정적 분석 도구가 잘못된 사용을 찾는 데 도움을 준다.

````python
print('--- 함수 type hint ---')

def add(a: int, b: int) -> int:
    return a + b

print('add(10, 20):', add(10, 20))

# 런타임에서는 type hint가 직접 타입을 막지 않는다.
# 정적 분석 도구는 아래 호출을 경고할 수 있다.
print("add('10', '20'):", add('10', '20'))
````

## 6. 상수 표현과 Final

Python에는 값을 절대 바꾸지 못하게 만드는 진짜 상수 문법이 따로 있는 것은 아니다.

- 관례상 대문자로 작성된 변수는 상수처럼 다룬다.
- `typing.Final`은 이 값이 재대입되지 않아야 한다는 힌트를 제공한다.
- `Final`도 정적 분석 도구를 위한 힌트이며, Python 런타임이 재대입을 직접 막지는 않는다.

````python
from typing import Final

print('--- Final 상수 표현 ---')

MAX_COUNT: Final[int] = 10
print('처음 MAX_COUNT:', MAX_COUNT)

# Final 역시 정적 분석 도구를 위한 힌트이며, 파이썬 런타임이 재대입을 직접 막지는 않는다.
MAX_COUNT = 5
print('재대입 후 MAX_COUNT:', MAX_COUNT)
````
