# match...case

- `match...case`는 값의 패턴에 따라 분기 처리하는 제어문이다.
- 다른 언어의 `switch...case`와 비슷하지만, Python의 `match...case`는 값뿐 아니라 구조 패턴도 검사할 수 있다.
- Python 3.10 이상에서 사용할 수 있다.

## 1. 기본 문법

```python
match 표현식:
    case 값1:
        실행문
    case 값2:
        실행문
    case _:
        기본 실행문
```

`case _`는 앞의 어떤 조건에도 맞지 않을 때 실행되는 기본 분기이다.

````python
def print_drink_price(num: str) -> None:
    match num:
        case '1':
            print('콜라는 1000원입니다.')
        case '2':
            print('사이다는 1500원입니다.')
        case '3':
            print('쥬스는 2000원입니다.')
        case _:
            print('잘못 입력하셨습니다.')

    print('😊이용해 주셔서 감사합니다.😊')

print('--- 음료 선택 예제 ---')
for num in ['1', '2', '3', '9']:
    print(f'선택값: {num}')
    print_drink_price(num)
````

## 2. 계산기 분기

연산자처럼 정해진 값 중 하나를 선택하는 경우 `match...case`를 사용하면 분기 구조가 깔끔해진다.

````python
def calculate(operand: str, a: int, b: int) -> None:
    match operand:
        case '+':
            print(f'{a} + {b} = {a + b}')
        case '-':
            print(f'{a} - {b} = {a - b}')
        case '*':
            print(f'{a} * {b} = {a * b}')
        case '//':
            print(f'{a} // {b} = {a // b}')
        case '%':
            print(f'{a} % {b} = {a % b}')
        case '/':
            print(f'{a} / {b} = {a / b}')
        case _:
            raise ValueError(f'연산자가 부적절합니다.: {operand}')

print('--- 계산기 예제 ---')
for operand in ['+', '-', '*', '//', '%', '/']:
    calculate(operand, 10, 3)

try:
    calculate('@', 10, 3)
except ValueError as error:
    print('오류:', error)
````

## 3. guard 조건

`case 값 if 조건` 형태를 사용하면 단순 값 비교보다 더 세밀한 조건을 줄 수 있다.

````python
def classify_number(n: int) -> None:
    match n:
        case n if n > 0:
            print(f'{n}: 양수입니다.')
        case n if n < 0:
            print(f'{n}: 음수입니다.')
        case _:
            print(f'{n}: 0입니다.')

print('--- guard 조건 예제 ---')
for n in [10, -5, 0]:
    classify_number(n)
````

## 4. tuple 패턴 검사

`match...case`는 값 하나뿐 아니라 튜플 구조도 검사할 수 있다. `_`는 해당 자리에 어떤 값이 와도 된다는 뜻이다.

````python
def inspect_tuple(value: tuple[int, int]) -> None:
    match value:
        case (1, 2):
            print('(1, 2)가 전달되었습니다.')
        case (1, k) if k < 0:
            print('(1, 음수)가 전달되었습니다.')
        case (1, _):
            print('(1, n)가 전달되었습니다.')
        case _:
            print('그 외의 경우')

print('--- tuple 패턴 검사 ---')
for value in [(1, 2), (1, -10), (1, 10), (3, 4)]:
    print(f'value={value} -> ', end='')
    inspect_tuple(value)
````

## 5. dict 패턴 검사

딕셔너리도 특정 key와 value 패턴을 기준으로 분기할 수 있다.

````python
def inspect_applicant(applicant: dict[str, object]) -> None:
    match applicant:
        case {'type': 'dev', 'career': career} if career >= 10:
            print('중견 개발자 지원입니다.')
        case {'type': 'dev'}:
            print('개발자 지원입니다.')
        case {'type': 'devops'}:
            print('운영/개발자 지원입니다.')
        case _:
            print('기타 지원입니다.')

print('--- dict 패턴 검사 ---')
applicants = [
    {'type': 'dev', 'career': 15},
    {'type': 'dev', 'career': 5},
    {'type': 'devops', 'career': 0},
    {'type': 'hr'},
]
for applicant in applicants:
    print(f'{applicant} -> ', end='')
    inspect_applicant(applicant)
````

## 6. match...case로 학점 계산하기

`if...elif...else`로 작성했던 학점 계산도 `match...case`와 guard 조건으로 표현할 수 있다.

````python
def get_grade_by_match(score: int) -> str:
    if not (0 <= score <= 100):
        raise ValueError('유효한 점수를 입력해주세요...')

    match score:
        case score if score >= 90:
            grade = 'A'
        case score if score >= 80:
            grade = 'B'
        case score if score >= 70:
            grade = 'C'
        case score if score >= 60:
            grade = 'D'
        case _:
            grade = 'F'

    return grade

print('--- match...case 학점 예제 ---')
for score in [95, 85, 75, 65, 55]:
    grade = get_grade_by_match(score)
    print(f'{score}점 -> {grade}등급')

# 직접 입력 실습용
# score: int = int(input('시험점수 입력: '))
# print(get_grade_by_match(score))
````
