## Getter/Setter 메서드
캡슐화된 클래스에서 속성을 제어하기 위한 메서드로써 기본적으로 다음 두 유형이 있다.
1. getter 속성의 값을 읽어내기 위한 메서드
2. setter 속성의 값을 설정하기 위한 메서드

## Python의 `@property`

Python에서는 Java식 `get_value()`, `set_value()` 메서드도 사용할 수 있지만, 더 Python다운 방식으로는 `@property`를 자주 사용한다.

`@property`를 사용하면 메서드로 값을 검증하면서도, 사용하는 쪽에서는 속성처럼 읽고 쓸 수 있다.
