# dict (Dictionary, 사전 자료형)

- `dict`는 하나의 요소를 `key: value` 형식으로 저장하는 자료형이다.
- 사전에서 단어를 찾으면 뜻을 알 수 있는 것처럼, `dict`에서는 `key`를 통해 `value`를 조회한다.
- `list`처럼 인덱스로 접근하는 것이 아니라, 의미 있는 이름인 `key`로 값을 찾는 구조이다.

## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.

```python
print('--- dict 생성과 key 규칙 ---')

dct = {
    'a': 10,
    'b': 20,
    'a': 100  # 중복된 key를 등록하면 마지막 값으로 덮어써진다.
}
print('dct:', dct, type(dct))

# tuple은 불변 자료형이므로 key로 사용할 수 있다.
point_map = {
    (10, 20): '서울 좌표'
}
print("point_map[(10, 20)]:", point_map[(10, 20)])

# list는 가변 자료형이므로 key로 사용할 수 없다.
# invalid_key_map = {[10, 20]: '서울 좌표'}  # TypeError: unhashable type: 'list'
```

## 2. 요소 조회

`dict`는 `key`를 이용해 `value`를 조회한다.

- `dict[key]`는 해당 `key`가 없으면 `KeyError`가 발생한다.
- `dict.get(key)`는 해당 `key`가 없으면 기본적으로 `None`을 반환한다.
- `dict.get(key, 기본값)`을 사용하면 없는 `key`에 대한 기본값을 직접 정할 수 있다.

```python
print('--- dict 조회 ---')

print("dct['a'], dct['b']:", dct['a'], dct['b'])

key = 'a'
print('dct[key]:', dct[key])

# print(dct['c'])  # KeyError: 'c'
print("dct.get('a'):", dct.get('a'))
print("dct.get('c'):", dct.get('c'))
print("dct.get('c', '값 없음'):", dct.get('c', '값 없음'))
```

## 3. 요소 추가, 수정, 제거

`dict`는 가변 자료형이므로 생성 후에도 요소를 추가, 수정, 제거할 수 있다.

- `dict[key] = value`로 요소를 추가하거나 수정한다.
- `update()`로 여러 요소를 한 번에 추가할 수 있다.
- `None`을 넣는 것은 값을 비우는 것이지, 요소를 삭제하는 것은 아니다.
- 요소 자체를 제거하려면 `pop()`이나 `del`을 사용한다.

```python
print('--- dict 추가와 수정 ---')

dct['c'] = 3
dct.update(d=4)  # key=value 형식으로 작성
dct.update({'e': 5, 'f': 6})
print('추가 후 dct:', dct)

print('--- dict value를 None으로 변경 ---')
dct['f'] = None
print("dct['f'] = None 후 dct:", dct)

print('--- dict item 제거 ---')
removed_value = dct.pop('f')
print("pop('f')로 제거된 값:", removed_value)
print("pop('f') 후 dct:", dct)

# dct.pop('f')  # 이미 제거된 key를 다시 pop하면 KeyError가 발생한다.
```

## 4. dict 생성 함수와 주요 메서드

`dict()` 함수로도 딕셔너리를 만들 수 있다.

- `keys()`는 모든 key를 반환한다.
- `values()`는 모든 value를 반환한다.
- `items()`는 `(key, value)` 형태의 쌍을 반환한다.
- 이 결과들은 리스트가 아니라 `dict_keys`, `dict_values`, `dict_items` 같은 view 객체이다.

```python
print('--- dict() 함수 ---')

dct2 = dict(name='홍길동', age=22)
print('dict(name=..., age=...):', dct2)

# key/value로 구성된 tuple을 list로 전달
dct3 = dict([('name', '신사임당'), ('age', 33)])
print("dict([('name', ...), ('age', ...)]):", dct3)

print('--- dict 주요 메서드 ---')
keys = dct3.keys()
values = dct3.values()
items = dct3.items()

print('dct3.keys():', keys, type(keys))
print('dct3.values():', values, type(values))
print('dct3.items():', items, type(items))
```

## 5. dict 반복 순회와 unpacking

`dict`를 반복문에 바로 넣으면 기본적으로 `key`가 하나씩 나온다.

- `for key in dct`는 `for key in dct.keys()`와 비슷하게 동작한다.
- `values()`를 사용하면 값만 순회할 수 있다.
- `items()`를 사용하면 `(key, value)` 쌍을 순회할 수 있다.
- `for key, value in dct.items()`처럼 unpacking하면 key와 value를 바로 나눠 받을 수 있다.

```python
print('--- dict 순회: key만 꺼냄 ---')
for key in dct3:
    print('key로 조회한 값:', f'{key} = {dct3[key]}')

print('--- dict keys 순회 ---')
for key in dct3.keys():
    print('keys()로 조회한 값:', f'{key} = {dct3[key]}')

print('--- dict values 순회 ---')
for value in dct3.values():
    print('value:', value)

print('--- dict items 순회 ---')
for item in dct3.items():
    print('item:', item, type(item))
```

```python
print('--- dict items unpacking ---')
for key, value in dct3.items():
    print('key, value:', key, value)

print('--- tuple unpacking ---')
abc = ('A', 'a')
m, n = abc
print('m, n:', m, n)
```

## 6. dict 대입, 얕은 복사, 깊은 복사

복사는 수업에서 학생들이 자주 헷갈리는 부분이다.

- **대입**: 같은 객체를 함께 바라보도록 참조를 복사한다.
- **얕은 복사(shallow copy)**: 새 컨테이너 객체를 만들지만, 내부 중첩 객체는 공유할 수 있다.
- **깊은 복사(deep copy)**: 내부 중첩 객체까지 재귀적으로 새로 복사한다.
- 중첩 객체가 없는 단순 `dict`에서는 `copy()`가 깊은 복사처럼 보일 수 있지만, 정확히는 얕은 복사이다.

```python
sample = {
    'name': '기계식 키보드',
    'price': 30_000,
    'origin': 'KR',
}

print('--- dict 대입: 같은 객체를 함께 참조 ---')
other = sample
print('id(sample), id(other):', id(sample), id(other))

print("before sample['name']:", sample['name'])
print("before other['name']:", other['name'])

sample['name'] = '멋진 키보드'

print("after sample['name']:", sample['name'])
print("after other['name']:", other['name'])
print('sample is other:', sample is other)
```

```python
print('--- dict 얕은 복사: copy() ---')

another = sample.copy()
print('id(sample), id(another):', id(sample), id(another))
print('sample is another:', sample is another)

sample['price'] *= 10
print('sample 변경 후 sample:', sample)
print('sample 변경 후 another:', another)
```

```python
print('--- 중첩 dict copy()와 deepcopy() 비교 ---')
import copy

nested_sample = {
    'name': '노트북',
    'options': ['16GB RAM', '512GB SSD']
}

shallow_copy = nested_sample.copy()
deep_copy = copy.deepcopy(nested_sample)

nested_sample['options'].append('외장 그래픽')
print('원본:', nested_sample)
print('copy() 결과:', shallow_copy)
print('deepcopy() 결과:', deep_copy)
```

## 7. list에서도 같은 복사 개념 확인하기

`dict`뿐 아니라 `list`도 대입과 `copy()`의 차이를 가진다.

- 대입하면 두 변수가 같은 리스트 객체를 바라본다.
- `copy()`를 사용하면 새 리스트 객체를 만든다.
- 단, 리스트 안에 또 다른 리스트나 딕셔너리가 들어 있으면 `copy()`도 얕은 복사라는 점을 다시 떠올려야 한다.

```python
print('--- list 대입/copy() 비교 ---')

prices = [10_000, 20_000, 30_000]

# [대입] prices와 prices_copy가 같은 list 객체를 함께 바라본다.
prices_copy = prices
prices.append(40_000)
print('대입 후 prices_copy:', prices_copy)

# [copy()] prices의 복사본을 만들어 각각 10% 값을 올려서 관리한다.
new_prices = prices.copy()
new_prices[0] *= 1.1
new_prices[1] *= 1.1
new_prices[2] *= 1.1
new_prices[3] *= 1.1
print('원본 prices:', prices)
print('copy() 후 수정한 new_prices:', new_prices)
```
