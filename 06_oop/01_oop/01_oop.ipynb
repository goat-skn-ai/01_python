
# OOP
Object-Oriented Programming 객체지향 프로그래밍
객체(메모리조각)를 중심으로 프로그램을 작성하는 방법론

- 변수(속성), 함수(메소드)를 묶어서 하나의 단위로써 프로그래밍
- 클래스를 작성후에 이를 바탕으로 n개의 객체(instance)를 생성할 수 있다.

```python
# 클래스 (객체의 설계도)
class Person:

    def __init__(self, name, age):
        """생성자 함수 (객체의 속성 초기화를 담당)"""
        # 현재 객체(self)의 속성을 선언
        self.name = name
        self.age = age

    def greeting(self):
        print(f'안녕하세요, 저는 {self.age}살 {self.name}입니다.')
```

```python
# 클래스와 객체(인스턴스)
hong = Person('홍길동', 33)
sinsa = Person('신사임당', 48)

print(hong, type(hong))
print(sinsa, type(sinsa))

print(hong.name, hong.age)
hong.greeting()

print(sinsa.name, sinsa.age)
sinsa.greeting()
```

## 속성/메소드

* 클래스 속성: 인스턴스간에 공유될 수 있는 속성
* 클래스 메소드: 인스턴스간에 공유될 수 있는 메소드

* 인스턴스 속성: 인스턴스마다 달라지는 속성
* 인스턴스 메소드: 인스턴스속성에 대해서 처리하는 메소드


```python
class SKNetworkStudent:
    # 클래스 속성
    course_name = 'Family AI Camp'

    def __init__(self, nth, name):
        # 인스턴스 속성
        self.nth = nth
        self.name = name

    # 인스턴스 메소드
    def greeting(self):
        print(f'안녕하세요. {self.nth}기 {self.name}입니다.')

    # 클래스 메소드
    @classmethod # 데코레이터
    def good_morning(cls):
        # 클래스 객체 사용가능
        print(f'{cls.course_name} 학생여러분, 좋은 아침입니다.')

    @staticmethod
    def good_bye(name):
        print(f'{SKNetworkStudent.course_name} {name}님, 오늘도 수고하셨습니다.')
```

```python
# 클래스 객체
print(SKNetworkStudent, type(SKNetworkStudent)) # 클래스객체

# 인스턴스 생성
kim = SKNetworkStudent(14, '김동현')
kong = SKNetworkStudent(14, '공지환')
sinsa = SKNetworkStudent(13, '신사임당')

print(kim, type(kim))
print(kong, type(kong))
print(sinsa, type(sinsa))
```

```python
# 클래스 속성/메소드
print(SKNetworkStudent.course_name)

SKNetworkStudent.good_morning()
SKNetworkStudent.good_bye('김광령')
```

```python
# 인스턴스 속성/메소드
print(kim.nth, kim.name)
kim.greeting()

print(kong.nth, kong.name)
kong.greeting()

print(sinsa.nth, sinsa.name)
sinsa.greeting()
```

## 객체지향 원칙
1. 캡슐화
  - 속성/메소드(함수) 묶어서 선언하고
  - 속성의 직접적인 접근을 막고, 메소드를 통한 사용만 허용.
2. 상속
  - 부모클래스에 선언된 속성/메소드를 자식클래스에서 선언없이 사용하는 것.
  - 코드 재사용성이 좋다.
3. 다형성
  - 상속에 기반한 기술. 여러 자식클래스를 부모타입으로 제어하는 것.
4. 추상화
  - 공통된, 보편적인 속성/메소드를 추출해 내는 것. 또는 이를 통해 제어
하는 것.
