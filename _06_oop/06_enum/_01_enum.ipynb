# Enum

- `Enum`은 서로 관련된 상수들을 하나의 이름 공간(namespace) 안에 묶어 관리하는 클래스이다.
- Python에서는 `enum.Enum`을 상속해서 Enum 클래스를 만든다.
- 문자열이나 숫자를 직접 비교하는 대신, 의미 있는 이름을 가진 상수로 비교할 수 있어 코드 가독성과 안정성이 좋아진다.
- 데이터 타입 단원에서는 심화/보충 주제로 가볍게 보고, 객체지향 수업에서는 클래스와 함께 다시 연결하면 좋다.


## 1. Enum 클래스 정의

`Enum`을 상속한 클래스 안에 `이름 = 값` 형태로 상수를 정의한다.

아래 예제에서는 성별을 표현하는 `Gender`, 크기를 표현하는 `Size` Enum을 만든다.


```python
from enum import Enum

print('--- enum 클래스 정의 ---')

class Gender(Enum):
    M = '남'
    F = '여'


class Size(Enum):
    S = 0
    M = 1
    L = 2


print(Gender.M)
print(Gender.F)
print(Size.M)
```

## 2. Enum을 사용하는 클래스

`Person` 객체가 성별 정보를 가질 때 문자열 `'M'`, `'F'`를 직접 저장하는 대신 `Gender.M`, `Gender.F`를 저장하도록 만들 수 있다.

이렇게 하면 허용된 값만 사용하도록 유도할 수 있고, 코드에서 값의 의미도 더 명확해진다.


```python
class Person:
    def __init__(self, name: str, gender: Gender) -> None:
        self.__name = name
        self.__gender = gender

    def __str__(self) -> str:
        return f'Person(name={self.__name}, gender={self.__gender})'

    def get_gender(self):
        return self.__gender
```

## 3. Enum 사용

객체를 만들 때 `Gender.M`, `Gender.F`처럼 Enum 멤버를 전달한다.


```python
print('--- enum 사용 ---')

hong = Person('홍길동', Gender.M)
sinsa = Person('신사임당', Gender.F)

print(hong)
print(sinsa)
```

## 4. Enum 객체와 타입

Enum 멤버는 단순 문자열이 아니라 Enum 클래스의 객체이다.


```python
print('--- enum 객체 ---')

print(Gender.M, type(Gender.M))
print(Gender.F, type(Gender.F))
```

## 5. name과 value

Enum 멤버는 `name`과 `value`를 가진다.

- `name`: Enum 안에서 사용하는 상수 이름
- `value`: 상수에 연결된 실제 값


```python
print('--- enum name/value ---')

print(Gender.M.name, type(Gender.M.name))
print(Gender.M.value, type(Gender.M.value))
```

## 6. Enum 비교

Enum은 문자열 값과 직접 비교하기보다는 Enum 멤버끼리 비교하는 것이 안전하다.


```python
print('--- enum 비교 ---')

print(hong.get_gender() == Gender.M)      # True
print(hong.get_gender() == 'M')           # False
print(hong.get_gender().name == 'M')      # True
```

## 7. Enum 순회

Enum 클래스는 `for`문으로 순회할 수 있다.


```python
print('--- enum 순회 ---')

for gender in Gender:
    print(gender)
```

## 8. 문자열에서 Enum으로 변환

사용자 입력이나 파일에서 읽어 온 값이 문자열일 때는 `Gender['M']`처럼 대괄호로 Enum 멤버를 찾을 수 있다.

노트북에서는 `input()`이 실행을 멈출 수 있으므로, 먼저 샘플 문자열로 동작을 확인한다.


```python
print('--- 문자열 to enum ---')

gender_input = 'M'
gender = Gender[gender_input]

print(gender)
print(gender.value)
```

## 8-1. 잘못된 문자열 처리

`Gender['M']`처럼 문자열로 Enum을 찾을 때, 존재하지 않는 이름을 사용하면 `KeyError`가 발생한다.
사용자 입력을 받을 때는 예외 처리를 함께 작성하는 것이 안전하다.


```python
print('--- 잘못된 enum 문자열 처리 ---')

wrong_gender_input = 'X'

try:
    gender = Gender[wrong_gender_input]
except KeyError as error:
    print('잘못된 성별 입력:', wrong_gender_input)
    print('사용 가능한 값:', list(Gender.__members__.keys()))
```

## 9. 사용자 입력으로 객체 생성하기

실제 프로그램에서는 `input()`으로 이름과 성별을 입력받아 `Person` 객체를 만들 수 있다.

수업 노트북에서는 실행 흐름이 멈추지 않도록 샘플값으로 같은 구조를 확인한다.


```python
print('--- 사용자 입력 예시를 샘플값으로 실행 ---')

name = '이몽룡'
gender_input = 'M'
gender = Gender[gender_input]
person = Person(name, gender)

print(person)
```

## 10. __members__ 확인

`__members__`를 사용하면 Enum에 정의된 멤버 목록을 확인할 수 있다.


```python
print('--- enum members ---')

print(Gender.__members__.keys())    # name 목록
print(Gender.__members__.values())  # Enum 멤버 목록
print(Gender.__members__.items())   # (name, Enum 멤버) 쌍
```
