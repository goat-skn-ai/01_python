# pip, requirements.txt, 가상환경 실제 설치 확인 실습

이 노트북은 `requirements.txt` 기반 설치 흐름을 설명하고, 현재 Python 환경에 필요한 패키지가 설치되어 있는지 확인한다.

가상환경은 프로젝트마다 독립된 Python 실행 환경과 패키지 설치 공간을 만들어 주는 기능이다.
전역 Python에 계속 설치하면 프로젝트별 패키지 버전이 충돌할 수 있으므로, 실무에서는 가상환경을 먼저 만들고 패키지를 설치하는 흐름을 권장한다.


## 1. 터미널 실습 흐름

아래 명령은 노트북 코드 셀에서 실행하는 것이 아니라, 터미널에서 순서대로 실행하는 실습 명령이다.


## 2. 설치 대상 패키지와 import 이름

PyPI에서 설치할 때 쓰는 패키지 이름과 Python 코드에서 `import`할 때 쓰는 모듈 이름이 다를 수 있다.
예를 들어 `python-dotenv`는 설치 이름이고, 코드에서는 `dotenv`로 import한다.


## 3. 설치 여부 확인 함수

`importlib.metadata.version()`으로 현재 환경에 특정 배포 패키지가 설치되어 있는지 확인할 수 있다.
설치되어 있지 않으면 `PackageNotFoundError`가 발생한다.


## 4. 설치 확인 실행

필요한 패키지가 설치되어 있지 않으면 설치 안내 명령을 출력한다.
모두 설치되어 있으면 버전 확인, import 확인, 간단한 사용 예시를 실행한다.
