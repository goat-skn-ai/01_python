## Getter/Setter 메서드
캡슐화된 클래스에서 속성을 제어하기 위한 메서드로써 기본적으로 다음 두 유형이 있다.
1. getter 속성의 값을 읽어내기 위한 메서드
2. setter 속성의 값을 설정하기 위한 메서드

```python
# class명은 관례적으로 파스칼케이싱을 따른다.
class MyClass:

    def __init__(self, value):
        self.__value = value

    # 읽기 가능한 속성인 경우, getter를 선언해둔다.
    def get_value(self):
        return self.__value

    # 쓰기 허용 속성이라면, setter를 선언해둔다.
    def set_value(self, value):
        self.__value = value


my_clz = MyClass(10)

# 값 읽기
# my_clz.__value # AttributeError: 'MyClass' object has no attribute '__value'
print(my_clz.get_value())

# 값 쓰기
# my_clz.__value = 123
my_clz.set_value(20)

print(my_clz.get_value())
```

## Python의 `@property`

Python에서는 Java식 `get_value()`, `set_value()` 메서드도 사용할 수 있지만, 더 Python다운 방식으로는 `@property`를 자주 사용한다.

`@property`를 사용하면 메서드로 값을 검증하면서도, 사용하는 쪽에서는 속성처럼 읽고 쓸 수 있다.


```python
print('--- @property 예제 ---')

class Score:
    def __init__(self, value: int):
        self.value = value

    @property
    def value(self):
        return self.__value

    @value.setter
    def value(self, value: int):
        if not 0 <= value <= 100:
            raise ValueError('점수는 0 이상 100 이하만 가능합니다.')

        self.__value = value

score = Score(80)
print('score.value 초기값:', score.value)

score.value = 95
print('score.value 변경 후:', score.value)

try:
    score.value = 120
except ValueError as error:
    print('잘못된 점수 대입 결과:', error)
```
