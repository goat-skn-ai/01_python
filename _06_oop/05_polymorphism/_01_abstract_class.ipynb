# 추상클래스 Abstract Class
- 자식클래스에서 구현해야 할 추상메소드를 가진 부모클래스 (인터페이스)
- 추상메소드는 자식클래스에서 반드시 구현해야할 의무를 가진다.
- 추상클래스는 직접 객체화 할수 없다.

- abc.ABC클래스를 상속하면 추상클래스를 선언할 수 있다.
- @abstractmethod 데코레이터로 추상메소드를 선언할 수 있다.

```python
# 추상 클래스 ABC import
# ABC(Abstract Base Class): 직접 객체를 만들기보다, 자식 클래스가 따라야 할 공통 규격을 정의할 때 사용
# abstractmethod: 자식 클래스에서 반드시 재정의해야 하는 메서드를 표시
from abc import ABC, abstractmethod


# Product는 모든 상품이 공통으로 가져야 할 상태와 기능을 정의하는 부모 클래스
# ABC를 상속했기 때문에 추상 클래스로 사용할 수 있다.
class Product(ABC):
    def __init__(self, code: str, name: str, price: int):
        self.__code = code
        self.__name = name
        self.__price = price

    # 추상메서드
    # 상품마다 세금 계산 방식이 달라질 수 있으므로, 부모 클래스에서는 규격만 정하고 구현은 자식에게 맡긴다.
    @abstractmethod
    def get_tax(self):
        pass

    def get_price(self):
        return self.__price

    def __str__(self):
        return f'code={self.__code}, name={self.__name}, price={self.__price}'


# TV는 Product의 구체 클래스(concrete class)
class TV(Product):
    def __init__(self, code: str, name: str, price: int, size: int):
        # 부모 클래스의 생성자를 먼저 호출해서 공통 속성(code, name, price)을 초기화한다.
        super().__init__(code, name, price)
        self.__size = size

    # Product의 추상메서드를 반드시 구현해야 TV 객체를 생성할 수 있다.
    # 여기서는 TV의 세금을 가격의 10%로 계산한다.
    def get_tax(self):
        return self.get_price() * 0.1

    # 부모의 문자열 표현에 TV 전용 속성(size)을 덧붙인다.
    def __str__(self):
        return f'TV({super().__str__()}, size={self.__size})'


# 추상클래스는 객체화할 수 없다.
# product = Product('1234', 'ㅋㅋㅋ', 100)


# get_tax 추상메소드를 구현하지 않으면...
# TypeError: Can't instantiate abstract class TV without an implementation for abstract method 'get_tax'

# 같은 TV 클래스라도 객체마다 상태(code, name, price, size)는 다르게 가질 수 있다.
tv = TV('samsung-tv-1234', '삼성 와이드8k TV', 10_000_000, 100)
print(tv)
print(tv.get_tax())

# 다형성 관점에서는 Product 타입의 규격(get_tax)을 믿고, 실제 계산은 자식 클래스(TV)의 구현에 맡긴다.
tv2 = TV('xiaomi-1234', '샤오미 TV', 5_000_000, 100)
print(tv2)
print(tv2.get_tax())
```
