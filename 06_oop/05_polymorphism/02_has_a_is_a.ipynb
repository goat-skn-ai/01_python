# Has-a 포함관계
- 하나의 클래스가 다른 클래스를 속성으로 가지고 있는 경우
- A has a B 성립하는 경우

```python
class Engine:
    def __init__(self, horse_power: int):
        self.horse_power = horse_power

    def start_engine(self):
        pass

    def stop_engine(self):
        pass

class Car:
    def __init__(self, name: str, engine: Engine):
        self.name = name
        self.engine = engine

    def drive(self):
        self.engine.start_engine()
        print('Driving...')
        self.engine.stop_engine()

# class Car(Engine):
#
#     def __init__(self, name: str):
#         self.name = name
#
#     def drive(self):
#         self.start_engine()
#         print('Driving...')
#         self.stop_engine()

```

# Is-a 상속관계
* 하나의 클래스가 다른 클래스를 상속하는 경우
* A is a B 성립하는 경우

```python
class Taxi(Car):
    def __init__(self, name: str, engine: Engine, type: str):
        super().__init__(name, engine)
        self.type = type # 개인/회사 택시 여부

taxi = Taxi('소나타', Engine(200), '개인')
taxi.drive()

class Bus(Car):
    pass

class Truck(Car):
    pass
```
