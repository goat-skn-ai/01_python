# Function 함수

- 함수는 반복되거나 절차적인 실행 코드를 하나의 이름으로 묶은 것이다.
- 함수를 선언한 뒤 호출해서 사용한다.
- 호출할 때 값을 전달할 수 있고, 실행 결과를 반환받을 수 있다.
- Python에서 함수는 하나의 값처럼 다룰 수 있는 1급 객체이다.

## 1. 함수 선언과 호출

`def` 키워드로 함수를 선언하고, 함수 이름 뒤에 괄호를 붙여 호출한다.

```python
def add(a, b):
    print(f'a = {a}, b = {b}')
    return a + b

print('add 함수 객체:', add, type(add))

result = add(10, 20)
print(f'정수 덧셈 result = {result}')

result = add('Hello', 'world')
print(f'문자열 연결 result = {result}')

result = add([1, 2, 3], ['🍕', '🍖', '☕'])
print(f'리스트 연결 result = {result}')
```

## 2. 매개변수와 인수

- **매개변수(parameter)**: 함수 선언부에서 값을 받기 위해 만든 변수
- **전달인자(argument)**: 함수 호출부에서 실제로 전달하는 값

```python
def foo(k, m):
    print(f'k = {k}, m = {m}')

print('--- 직접 값 전달 ---')
foo(3, 4)

print('--- 변수에 담긴 값 전달 ---')
i, j = 10, 20
foo(i, j)
```

## 3. return과 함수 객체

함수는 실행을 마치면 호출한 위치로 돌아온다. 이때 `return`으로 값을 하나 돌려줄 수 있다.

또한 Python에서 함수는 변수에 담거나 다른 함수의 인수로 전달할 수 있다.

```python
def bar():
    print('bar')
    return 0

print('bar() 반환값:', bar())

f = bar
print('bar와 f:', bar, f, type(f))
print('bar is f:', bar is f)


def runner(func):
    """함수 실행기"""
    func()

runner(bar)
```

## 4. 위치 인수와 키워드 인수

함수를 호출할 때 인수를 전달하는 방식은 크게 두 가지이다.

- 위치 인수: 작성한 순서대로 매개변수에 전달
- 키워드 인수: `key=value` 형태로 같은 이름의 매개변수에 전달

두 방식을 섞을 때는 위치 인수를 먼저 쓰고, 그 뒤에 키워드 인수를 써야 한다.

```python
def test(a, b):
    print(f'a = {a}, b = {b}')

print('--- 위치 인수 ---')
test(10, 20)

print('--- 키워드 인수 ---')
test(a=1, b=2)
test(b=2, a=1)

print('--- 위치 인수와 키워드 인수 혼용 ---')
test(100, b=200)

# test(a=1, b=2, c=3)  # TypeError: 예상하지 못한 키워드 인수
# test(b=200, 100)     # SyntaxError: 키워드 인수 뒤에 위치 인수를 작성할 수 없음
```

## 5. 매개변수 기본값

기본값이 있는 매개변수는 호출할 때 생략할 수 있다.

단, 기본값이 있는 매개변수는 기본값이 없는 매개변수보다 오른쪽에 작성해야 한다.

```python
def test2(name, age=20, married=False):
    print(f'name = {name}, age = {age}, married = {married}')

print('--- 기본값 사용 ---')
test2('홍길동', 33)
test2('홍길동', 33, True)
test2('신사임당')

# def wrong(name, age=20, married):
#     pass
# SyntaxError: 기본값 없는 매개변수가 기본값 있는 매개변수 뒤에 올 수 없음
```

## 6. packing: `*args`, `**kwargs`

- `*args`: 여러 위치 인수를 하나의 튜플로 모은다.
- `**kwargs (keyword arguments)`: 여러 키워드 인수를 하나의 딕셔너리로 모은다.
- 매개변수 작성 순서는 일반 매개변수, `*args`, `**kwargs` 순서를 기본으로 생각하면 좋다.

```python
def test3(*args):
    print('args:', args, type(args))
    return sum(args)

print(test3(1, 2))
print(test3(1, 2, 3, 4, 5))


def test4(emoji, *args):
    print(f'emoji = {emoji}, args = {args}')

test4('☕', 1, 2, 3)
test4('☕', 10, 20)
test4('🍖')


def test5(**kwargs):
    print('kwargs:', kwargs, type(kwargs))

test5()
test5(name='윤봉길', age=44, hobby=['축구', '풋볼', '사커'])


def test6(name, age, **kwargs):
    print('name, age, kwargs:', name, age, kwargs, type(kwargs))

test6('윤봉길', 44, hobby=['축구', '풋볼', '사커'])


def test7(a, b, c, *args, **kwargs):
    print('a, b, c, args, kwargs:', a, b, c, args, kwargs)

test7(10, 20, 30, 40, 50, name='다비드', age=44)
```

## 7. unpacking: `*list`, `**dict`

- `*list`는 리스트의 요소를 꺼내어 위치 인수처럼 나열한다.
- `**dict`는 딕셔너리의 key-value를 꺼내어 키워드 인수처럼 나열한다.

```python
names = ['홍길동', '이순신', '유관순']


def test8(*args):
    print('args:', args)

print('--- list unpacking ---')
test8(names) # []로 묶인 상태로 전달
test8(names[0], names[1], names[2]) # 직접 list를 풀어서 각 요소를 전달
test8(*names) # list를 풀어서 각 요소를 전달



info = {'date': '2025/04/14', 'time': '16:45'}

def test9(date, time):
    print('date, time:', date, time)

print('--- dict unpacking ---')
test9(date='2025/04/14', time='16:45')
test9(date=info['date'], time=info['time'])
test9(**info)
```

## 8. 위치 전용과 키워드 전용 매개변수

Python 함수는 특정 매개변수를 위치 전용 또는 키워드 전용으로 제한할 수 있다.

- `/` 앞의 매개변수는 위치 전용이다.
- `*` 뒤의 매개변수는 키워드 전용이다.

```python
def test10(name, /, greeting='안녕'):
    print(name, greeting)

test10('길동아')
# test10(name='길동')  # TypeError: name은 위치 전용 매개변수


def test11(a, *, b, c):
    print(a, b, c)

# test11(10, 20, 30)  # TypeError: b, c는 키워드 전용 매개변수
test11(10, b=20, c=30)
```

## 9. 함수 type hint

함수의 매개변수와 반환값에도 type hint를 작성할 수 있다. type hint는 실행을 강제하기보다 IDE와 정적 분석 도구가 코드를 이해하도록 돕는 역할을 한다.

```python
def test12(a: str, b: int) -> str:
    return a + str(b)

print('test12("안녕", 123):', test12('안녕', 123))
```
