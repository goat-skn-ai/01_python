# PyPI, pip, requirements.txt

Python 프로젝트에서 외부 패키지를 설치하고 관리할 때 자주 만나는 세 가지 개념이다.

- PyPI(Python Package Index): Python 외부 패키지를 올리고 내려받는 공식 저장소
- pip: PyPI 같은 패키지 저장소에서 패키지를 설치, 삭제, 조회하는 도구
- requirements.txt: 프로젝트에 필요한 패키지 목록을 적어두는 의존성 명세 파일

pip는 설치 도구이고, 프로젝트 규모에 따라 `venv`, `requirements.txt`, `poetry`, `uv` 등을 함께 사용할 수 있다.


## 1. requirements.txt 예시

`requirements.txt`에는 설치할 패키지 이름과 버전 조건을 적는다.
빈 줄과 `#`으로 시작하는 주석은 설치 대상이 아니다.


## 2. 버전 조건 해석

패키지 버전을 고정하거나 범위를 지정하면 수업 환경과 실습 결과를 더 안정적으로 맞출 수 있다.


## 3. pip 명령어 흐름

| 명령어 | 의미 |
|---|---|
| `python3 -m venv .venv` | 현재 폴더에 `.venv`라는 이름의 파이썬 가상환경을 생성한다. |
| `source .venv/bin/activate` | 생성한 `.venv` 가상환경을 활성화한다. macOS/Linux 기준 명령어이다. |
| `python -m pip --version` | 현재 활성화된 파이썬 환경에서 사용하는 `pip` 버전을 확인한다. |
| `python -m pip install requests` | 현재 파이썬 환경에 `requests` 패키지를 설치한다. |
| `python -m pip show requests` | 설치된 `requests` 패키지의 버전, 위치, 의존성 등의 정보를 확인한다. |
| `python -m pip freeze > requirements.txt` | 현재 환경에 설치된 패키지 목록을 `requirements.txt` 파일로 저장한다. |
| `python -m pip install -r requirements.txt` | `requirements.txt` 파일에 적힌 패키지들을 현재 환경에 한 번에 설치한다. |
| `python -m pip uninstall requests` | 현재 파이썬 환경에서 `requests` 패키지를 삭제한다. |

## 4. 직접 실행 여부 확인

`.py` 파일에서는 `if __name__ == '__main__'` 구문을 사용해 직접 실행된 경우에만 특정 코드를 실행할 수 있다.
노트북에서는 보통 `__name__`이 `'__main__'`으로 동작한다.
