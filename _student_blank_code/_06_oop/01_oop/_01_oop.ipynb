
# OOP
Object-Oriented Programming 객체지향 프로그래밍
객체(메모리조각)를 중심으로 프로그램을 작성하는 방법론

- 변수(속성), 함수(메소드)를 묶어서 하나의 단위로써 프로그래밍
- 클래스를 작성후에 이를 바탕으로 n개의 객체(instance)를 생성할 수 있다.

## 속성/메서드

* 클래스 속성: 인스턴스간에 공유될 수 있는 속성
* 클래스 메서드: 인스턴스간에 공유될 수 있는 메소드

* 인스턴스 속성: 인스턴스마다 달라지는 속성
* 인스턴스 메서드: 인스턴스속성에 대해서 처리하는 메소드


## 객체지향 원칙
1. 캡슐화
  - 속성/메소드(함수) 묶어서 선언하고
  - 속성의 직접적인 접근을 막고, 메소드를 통한 사용만 허용.
2. 상속
  - 부모클래스에 선언된 속성/메소드를 자식클래스에서 선언없이 사용하는 것.
  - 코드 재사용성이 좋다.
3. 다형성
  - 상속에 기반한 기술. 여러 자식클래스를 부모타입으로 제어하는 것.
4. 추상화
  - 공통된, 보편적인 속성/메소드를 추출해 내는 것. 또는 이를 통해 제어
하는 것.
