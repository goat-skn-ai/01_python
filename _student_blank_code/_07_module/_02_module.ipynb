# 모듈 Module

모듈은 Python 코드를 담고 있는 `.py` 파일이다.
변수, 함수, 클래스 등을 모듈 단위로 분리하면 코드를 재사용하고 관리하기 쉽다.

- 표준 라이브러리 모듈: Python에 기본 포함된 모듈 예) `math`
- 사용자 정의 모듈: 직접 만든 `.py` 파일 예) `skn/my_math.py`
- `_` 또는 `__`로 시작하는 이름은 내부용이라는 관례가 있다.


## 1. 표준 라이브러리 모듈 import

`import math`는 Python 표준 라이브러리의 `math` 모듈을 가져온다.
`dir()`을 사용하면 모듈이나 현재 namespace에서 사용할 수 있는 이름을 확인할 수 있다.


## 2. `__name__` 확인

`__name__`은 현재 파일이 직접 실행 중인지, 다른 파일에서 import 되었는지 확인할 때 자주 사용하는 특별 변수이다.
직접 실행된 파일의 `__name__`은 `'__main__'`이 된다.


## 3. 사용자 정의 모듈 import

`skn` 폴더는 `__init__.py`가 있는 패키지이고, `my_math.py`는 그 안에 있는 사용자 정의 모듈이다.
모듈을 namespace로 사용하면 같은 이름의 변수 충돌을 줄일 수 있다.


## 4. import 방식과 내부용 이름

`from 모듈 import *`는 모듈 안의 여러 이름을 현재 namespace로 가져온다.
다만 `_`로 시작하는 이름은 내부용으로 간주되어 `import *` 대상에서 제외된다.
실제 수업에서는 이름 충돌을 줄이기 위해 `import 모듈명` 또는 `import 모듈명 as 별칭` 형태를 우선 권장한다.


## 5. `if __name__ == '__main__'`

아래 조건문은 파일을 직접 실행할 때만 특정 코드를 실행하고 싶을 때 사용한다.
다른 파일에서 import할 때 실행되면 안 되는 테스트 코드나 예제 코드를 보호할 수 있다.
