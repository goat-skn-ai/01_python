# 캡슐화
- 속성/메소드를 하나의 클래스에 함께 선언
- private 속성(변수) 선언해서 외부에서 직접 접근을 막고, 제공된 메소드를 통해서 사용하도록 강제
- 캡슐화 원칙을 잘 지켰다면, 주요한 데이터가 외부에 노출되지 않은 은닉화 효과까지 얻게된다.

```python
class Account:
    def __init__(self, username, balance):
        # private 접근제한 처리
        self.__username = username
        self.__balance = balance

    def info(self):
        print(f'{self.__username}님의 현재잔액은 {self.__balance}원 입니다.')

    def deposit(self, money):
        """
        입금메소드
        - money 유효성 검사
        - 오염된 입금처리 방지
        """
        if money <= 0:
            print('입금액이 유효하지 않습니다.')
        else:
            self.__balance += money
            print('입금이 정상처리 되었습니다.')

    def withdraw(self, money):
        """출금메소드"""
        if money <= 0:
            print('출금액이 유효하지 않습니다.')
        elif money > self.__balance:
            print('잔액이 부족합니다.')
        else:
            self.__balance -= money
            print('출금이 정상처리 되었습니다.')
```

```python
# private 속성 접근
acc = Account('honggd', 1_000_000)
# acc.name # AttributeError: 'Account' object has no attribute 'name'
# acc.__name # AttributeError: 'Account' object has no attribute '__name'
# acc.balance # AttributeError: 'Account' object has no attribute 'balance'
# acc.__balance # AttributeError: 'Account' object has no attribute '__balance'
```

```python
# 홍길동의 계좌관리
acc = Account('honggd', 1_000_000)
print(acc)
acc.info()

# 입금 처리
# acc.balance = acc.balance + 500_000
# acc.balance = acc.balance * 500_000
acc.balance = acc.balance + -500_000
acc.info()

# 출금 처리
acc.balance = acc.balance - 700_000
acc.info()
```

```python
# 홍길동의 계좌관리 (캡슐화)
acc = Account('honggd', 1_000_000)
print(acc)
acc.info()

# 입금처리
acc.deposit(500_000)
acc.deposit(-500_000)
acc.info()

# 출금처리
acc.withdraw(700_000)
acc.withdraw(-700_000)
acc.withdraw(40_000_000)
acc.info()
```
