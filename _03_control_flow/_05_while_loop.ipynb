# while 반복문

- `while` 반복문은 조건식이 `True`로 평가되는 동안 코드를 반복 실행한다.
- 조건식이 `False`가 되면 반복을 중지한다.
- 반복 횟수가 미리 정해져 있지 않고, 특정 조건이 만족될 때까지 반복해야 할 때 자주 사용한다.

## 1. 기본 while 구조

`while`은 보통 초기값, 조건식, 증감식을 함께 사용한다.

```python
i = 1
while i <= 3:
    print(i)
    i += 1
```

````python
print('--- 기본 while 반복문 ---')

i = 1
while i <= 3:
    print(i)
    i += 1
````

## 2. break로 반복 중지하기

`while True`는 조건식이 항상 참인 반복문이다. 이 구조를 사용할 때는 반드시 반복을 빠져나올 수 있는 `break` 조건을 만들어야 한다.

````python
print('--- while True와 break ---')

i = 1
while True:
    print(i)
    i += 1

    if i > 10:
        break

print('🍕🍕🍕🍕🍕🍕🍕')
````

## 3. 메뉴 주문 예제

실제 수업에서는 `input()`으로 사용자의 선택을 받을 수 있다. 노트북 기본 예제는 전체 실행이 멈추지 않도록 미리 준비한 선택값으로 시뮬레이션한다.

````python
menu = """
==============
SK 구내식당 메뉴
==============
1. 된장찌개
2. 비빔밥
3. 김치찌개
0. 종료
==============
선택: """

print(menu)

order: list[str] = []
sample_choices = ['1', '2', '9', '3', '0']
choice_index = 0

while True:
    choice = sample_choices[choice_index]
    choice_index += 1
    print('선택:', choice)

    match choice:
        case '1':
            order.append('된장찌개')
        case '2':
            order.append('비빔밥')
        case '3':
            order.append('김치찌개')
        case '0':
            break
        case _:
            print('잘못 입력하셨습니다.')

print(f'주문내역은 {order}입니다.')

# 직접 입력 실습용
# choice = input(menu)
````
