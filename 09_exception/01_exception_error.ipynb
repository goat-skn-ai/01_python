## 예외처리
**(중요) 발생된 예외/오류 상황에서 try..except구문을 통해 적절히 처리후 정상적인 흐름을 돌려내는 것**

```python
try:
    # 예외/오류가 발생할 수 있는 코드 작성
    # - 오류가 발생하면 except 블럭으로 즉시 이동 (이하 코드 실행 안함)
    # print(10 / 0)
    # - 오류가 발생하지 않으면, except구문은 무시하고 try블럭을 탈출!
    print(10 / 1)
    print('try블럭 실행 이상무!!!')
except:
    # 예외/오류가 발생했을때 처리할 코드/조치 작성
    print('안심하세요. 전 죽지 않아요....')

print('😊😊😊')
```

```python
# except블럭 여러가지 예외상황을 각각 처리가능
import random

try:
    # 난수 생성 0, 1
    n = random.randint(0, 1) # 시작값/종료값(포함) 사이의 정수 반환
    if n:
        print('abc' + 123) # TypeError 발생 코드
    else:
        print(int('abc')) # ValueError 발생 코드
except TypeError as e:
    print('TypeError가 발생했습니다!!', e, type(e))
except ValueError as e:
    print('ValueError가 발생했습니다!!', e, type(e))

print('😊😊😊')
```

```python
students: dict[int, dict] = {
    1: {"id": 1, "name": "홍길동"},
    2: {"id": 2, "name": "신사임당"},
    3: {"id": 3, "name": "이순신"},
}

# 학생 조회
# - 사용자로 부터 id값을 입력받아 해당 학생정보를 출력
# - 존재하지 않는 id를 입력해도 프로그램이 종료되어서는 안된다. "해당 id 학생은 존재하지 않습니다."
try:
    id = int(input("학생 ID 입력 : "))
    print(students[id]['name'])
except KeyError as e:
    print('해당 ID 학생은 존재하지 않습니다.')

print("😊😊😊")

```

```python
# 예외 무시 (주의)
try:
    print(10 / 0)
except Exception as e:
    # try블럭의 예외 추적 불가
    # pass
    # 예외로그 해두면 디버깅에 유리
    print(e)
```
