# lambda 람다

- `lambda`는 함수를 한 줄로 간결하게 선언하는 문법이다.
- 이름이 없는 작은 함수를 만들 때 사용한다.
- 기본 구조는 `lambda 매개변수: 반환값`이다.
- 복잡한 로직은 일반 `def` 함수로 작성하는 편이 읽기 쉽다.

## 1. lambda 기본 문법

아래 두 코드는 같은 의미이다.

```python
def square(x):
    return x ** 2

square = lambda x: x ** 2
```

## 2. list comprehension 안에서 사용하기

`lambda`는 짧은 계산식을 값처럼 전달하거나 즉시 실행할 때 사용할 수 있다.

## 3. 여러 매개변수 사용

`lambda`도 일반 함수처럼 하나 이상의 매개변수를 받을 수 있다.

## 4. 기본값, `*args` 사용

`lambda`에서도 기본값이 있는 매개변수와 `*args`를 사용할 수 있다. 다만 코드가 길어지면 `def` 함수로 분리하는 것이 좋다.
