## 요구사항

```
===== 위시리스트 관리 프로그램 =====
1. 상품 추가
2. 상품 삭제
3. 위시리스트 확인
4. 프로그램 종료
==========================
메뉴를 선택하세요:
```
* 사용자는 위 메뉴를 통해 본인의 위시리스트를 관리할 수 있습니다.
* 위 메뉴는 메뉴객체에 의해 제공되어야 합니다.
* 사용자 위시리스트 상품은 상품명, 가격 정보를 가질 수 있습니다.
* 위시리스트는 여러개의 상품을 관리할 수 있습니다. (list타입)
* `1번 상품추가`를 선택하면, 사용자에게 상품명과 가격을 각각 입력 받고, 상품 객체를 생성합니다. 이 상품 객체는 위시리스트에서 관리되어야 합니다.
* `2번 상품삭제`를 선택하면, 위시리스트에서 저장된 상품 인덱스를 추가로 입력받아 해당 상품을 삭제할 수 있습니다.
* `3번 위시리스트확인`을 선택하면 현재 위시리스트에 저장된 상품 목록을 출력합니다.
* `4번 프로그램 종료`를 선택하면 프로그램을 종료합니다.

아래 다이어그램을 참조할 수 있습니다.

```
## 객체 다이어그램 (Menu → WishList → Product)

+--------------+        +--------------+        +----------------+
|     Menu     |        |   WishList   |        |    Product     |
+--------------+        +--------------+        +----------------+
| - __wishlist | 1 ---->| - __products | <> ---->| - name: str     |
+--------------+        +--------------+        | - price: int    |
| + display()  |        | + add_product()       +----------------+
|              |        | + remove_product()
|              |        | + print_products()
+--------------+        +--------------+

```


````python
class Product:
    def __init__(self, name: str, price: int):
        # Product 객체가 알고 있는 것(속성)
        self.__name = name
        self.__price = price

    def get_name(self):
        return self.__name

    def get_price(self):
        return self.__price

    def __str__(self):
        return f'{self.__name}({self.__price}원)'

    def __repr__(self):
        # 가능하면 객체를 다시 만들 수 있는 형태의 문자열을 반환한다.
        return f"Product({self.__name!r}, {self.__price!r})"


class WishList:
    def __init__(self):
        # WishList 객체가 알고 있는 것(속성)
        self.__products: list[Product] = []

    def add_product(self, product: Product):
        self.__products.append(product)
        print('상품 추가 결과:', f'{product} 추가 완료')

    def remove_product(self, index: int):
        if not 0 <= index < len(self.__products):
            print('상품 삭제 결과:', '유효하지 않은 인덱스입니다.')
            return None

        return self.__products.pop(index)

    def print_products(self):
        print('--- 위시리스트 상품 목록 ---')

        if not self.__products:
            print('등록된 상품이 없습니다.')
            return

        for i, product in enumerate(self.__products):
            print(f'{i}: {product.get_name()} {product.get_price()}원')


class Menu:
    def __init__(self):
        self.__wishlist: WishList = WishList()

    def display(self, inputs: list[str] | None = None):
        menu = """
===== 위시리스트 관리 프로그램 =====
1. 상품 추가
2. 상품 삭제
3. 위시리스트 확인
4. 프로그램 종료
==========================
메뉴를 선택하세요: """
        input_iter = iter(inputs) if inputs is not None else None

        def read(prompt: str) -> str:
            if input_iter is None:
                return input(prompt)

            value = next(input_iter)
            print(prompt, value)
            return value

        while True:
            try:
                choice = read(menu)
            except StopIteration:
                choice = '4'
                print(menu, choice)

            match choice:
                case '1':
                    name = read('상품명 입력:')
                    price = int(read('가격 입력:'))
                    product = Product(name, price)
                    self.__wishlist.add_product(product)
                case '2':
                    self.__wishlist.print_products()
                    index = int(read('삭제할 상품의 인덱스를 입력해주세요:'))
                    removed = self.__wishlist.remove_product(index)
                    if removed:
                        print('상품 삭제 결과:', f'{removed} 삭제 완료')
                case '3':
                    self.__wishlist.print_products()
                case '4':
                    break
                case _:
                    print('잘못 입력하셨습니다.')

        print('이용해 주셔서 감사합니다.')
````

````python
# 실제 수업에서는 menu.display()로 실행하면 input()을 받을 수 있다.
# 노트북 자동 실행 검증을 위해 여기서는 샘플 입력을 전달한다.
menu = Menu()
menu.display(['1', 'ABC 초콜렛', '1000', '1', '노트북 파우치', '25000', '3', '2', '0', '3', '4'])
````

````python
prod = Product('ABC 초콜렛', 1000)

print('str(prod):', str(prod))
print('repr(prod):', repr(prod))
print('print(prod) 결과:', prod)
````
