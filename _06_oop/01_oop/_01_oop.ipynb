
# OOP
Object-Oriented Programming 객체지향 프로그래밍
객체(메모리조각)를 중심으로 프로그램을 작성하는 방법론

- 변수(속성), 함수(메소드)를 묶어서 하나의 단위로써 프로그래밍
- 클래스를 작성후에 이를 바탕으로 n개의 객체(instance)를 생성할 수 있다.

```python
# 클래스 (인스턴스(객체)를 만들기 위한 객체 == 인스턴스의 설계도)
class Person:

    nationality = 'Korea'

    def __init__(self, name, age):
        """생성자 함수 (객체의 속성 초기화를 담당)"""

        # 현재 객체(self)의 속성을 선언
        self.name = name
        self.age = age

    def greeting(self):
        print(f'안녕하세요, 저는 {self.age}살 {self.name}입니다.')
```

## 생성자 `__init__`

생성자는 객체가 생성될 때 자동으로 실행되어, 객체가 처음 가져야 할 속성을 초기화하는 메서드이다.

Python에서는 생성자 역할을 하는 메서드를 `__init__`이라는 이름으로 정의한다.
아래 코드에서 `Person('홍길동', 33)`을 실행하면 `__init__(self, name, age)`가 자동으로 호출되고, 전달한 값이 `self.name`, `self.age`에 저장된다.

- `self`: 생성된 인스턴스 자신
- `name`, `age`: 객체를 만들 때 외부에서 전달받는 값
- `self.name = name`: 전달받은 `name` 값을 현재 객체의 속성으로 저장

즉, 생성자는 객체가 만들어지는 순간 필요한 초기 데이터를 세팅하는 초기화 담당 메서드이다.


```python
# 클래스와 인스턴스
hong = Person('홍길동', 33)
sinsa = Person('신사임당', 48)

print(hong, type(hong))
print(sinsa, type(sinsa))

print(hong.name, hong.age)
hong.greeting()

print(sinsa.name, sinsa.age)
sinsa.greeting()
```

## self 이해하기

`self`는 메서드를 호출한 인스턴스 자신을 가리키는 매개변수이다.

아래 두 호출은 같은 의미로 볼 수 있다.

- `hong.greeting()`
- `Person.greeting(hong)`

일반적으로는 첫 번째 방식처럼 인스턴스를 통해 메서드를 호출한다.


```python
print('--- self 이해하기 ---')

hong = Person('홍길동', 33)

print('hong.greeting() 호출:')
hong.greeting()

print('Person.greeting(hong) 호출:')
Person.greeting(hong)
```

```python
class SKNetworkStudent:
    # 클래스 속성
    # 모든 인스턴스가 함께 공유하는 값
    course_name = 'Family AI Camp'

    def __init__(self, nth, name):
        # 인스턴스 속성
        # 객체(인스턴스)마다 개별적으로 가지는 값
        self.nth = nth
        self.name = name

    # 인스턴스 메서드
    # 첫 번째 매개변수로 self를 받는다.
    # self는 메서드를 호출한 인스턴스 자신을 의미한다.
    def greeting(self):
        print(f'안녕하세요. {self.nth}기 {self.name}입니다.')

    # 클래스 메서드
    # @classmethod 데코레이터를 붙인다.
    # 첫 번째 매개변수로 cls를 받는다.
    # cls는 클래스 자신, 여기서는 SKNetworkStudent를 의미한다.
    @classmethod
    def good_morning(cls):
        # cls를 통해 클래스 속성에 접근할 수 있다.
        print(f'{cls.course_name} 여러분, 좋은 아침입니다.')

    # 정적 메서드
    # @staticmethod 데코레이터를 붙인다.
    # self나 cls를 자동으로 받지 않는다.
    # 인스턴스 상태나 클래스 상태를 직접 다루지 않는 독립적인 기능에 사용한다.
    @staticmethod
    def good_bye(name):
        # staticmethod는 cls를 받지 않으므로,
        # 클래스 속성을 쓰려면 클래스명을 직접 적어야 한다.
        print(f'{SKNetworkStudent.course_name} {name}님, 오늘도 수고하셨습니다.')
```

```python
# 클래스 객체(class도 객체다)
print(SKNetworkStudent, type(SKNetworkStudent)) # 클래스객체

# 인스턴스 생성(주소 다름 확인)
kim = SKNetworkStudent(14, '김동현')
kong = SKNetworkStudent(14, '공지환')
sinsa = SKNetworkStudent(13, '신사임당')

print(kim, type(kim))
print(kong, type(kong))
print(sinsa, type(sinsa))
```

```python
# 인스턴스 생성
student = SKNetworkStudent(33, '김광령')

# 인스턴스 속성/메서드
print(student.nth)
print(student.name)
student.greeting()

# 클래스 속성
print(student.course_name)
print(SKNetworkStudent.course_name) # 클래스명.속성명 직접 작성 가능


# 인스턴스 메서드는 보통 인스턴스를 통해 호출한다.
# SKNetworkStudent.greeting()  # TypeError: self가 전달되지 않음


# 클래스 메서드는 cls를 통해 클래스 상태를 사용하고,
#정적 메서드는 self나 cls 없이 클래스 안에 넣어둔 독립 함수처럼 동작한다.

# 클래스 메서드는 클래스를 통해 호출할 수 있다.
SKNetworkStudent.good_morning()

# 정적 메서드도 클래스를 통해 호출할 수 있다.
SKNetworkStudent.good_bye('김광령')
```

## 클래스 속성과 인스턴스 속성 주의점

- 클래스 속성은 모든 인스턴스가 공유하는 값이다.
- 반면 인스턴스에 같은 이름의 속성을 새로 대입하면, 그 인스턴스만의 속성이 생겨 클래스 속성을 가리게 된다.


```python
print('--- 클래스 속성과 인스턴스 속성 ---')

kim = SKNetworkStudent(33, '김학생')
lee = SKNetworkStudent(33, '이학생')

print('변경 전 kim.course_name:', kim.course_name)
print('변경 전 lee.course_name:', lee.course_name)

SKNetworkStudent.course_name = 'SKN AI Camp'
print('클래스 속성 변경 후 kim.course_name:', kim.course_name)
print('클래스 속성 변경 후 lee.course_name:', lee.course_name)

# 인스턴스 속성을 추가(클래스 속성명과 같은 이름으로 추가)
kim.course_name = '개인별 과정명'
print('kim 인스턴스 속성 추가 후 kim.course_name:', kim.course_name)
print('kim 인스턴스 속성 추가 후 lee.course_name:', lee.course_name)
print('kim.__dict__:', kim.__dict__)
```

## 객체지향 원칙
1. 캡슐화
  - 속성/메소드(함수) 묶어서 선언하고
  - 속성의 직접적인 접근을 막고, 메소드를 통한 사용만 허용.
2. 상속
  - 부모클래스에 선언된 속성/메소드를 자식클래스에서 선언없이 사용하는 것.
  - 코드 재사용성이 좋다.
3. 다형성
  - 상속에 기반한 기술. 여러 자식클래스를 부모타입으로 제어하는 것.
4. 추상화
  - 공통된, 보편적인 속성/메소드를 추출해 내는 것. 또는 이를 통해 제어
하는 것.
